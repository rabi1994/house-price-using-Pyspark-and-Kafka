```python
from sklearn.model_selection import train_test_split
X=data.drop(['price','date'],axis=1)
y=data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_squared_error
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predections=model.predict(X_test)
```

```python
print(mean_squared_error(predections,y_test)**0.5)
```

```python
print(model.score(X_test,y_test))
```

```python
pred=pd.DataFrame(predections,columns=['predections'])
```

```python
all_scores=pd.concat([pred,y_test],axis=1)
```

# modeling and predictions 

```python
X=new_data.drop(['price','date'],axis=1)
y=new_data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,predictions)**2)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.linear_model import ElasticNet 
model=ElasticNet(alpha=0.03)
```

```python
model.fit(X_train,y_train)
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,predictions))
```

```python
model.score(X_test,y_test)
```
